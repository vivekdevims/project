```python
import random
subject = ["Elon Musk",
           "Salman Khan",
           "Virat Kohli",
           "Priyanka Chopra"]

Object = ["started dancing",
          "was found eating noodles",
          "was caught flying kites",
          "launched AI for cats",
          "was seen riding a cycle",
          "opened a tea stall",
          "was seen riding an auto-rickshaw",
          "was found giving a speech to a group of dogs"]

place = ["in Central Park",
         "in Delhi",
         "at India Gate",
         "in Pune",
         "on the streets of Mumbai",
         "at the railway station"]

while True:
  sub=random.choice(subject)
  obj=random.choice(Object)
  pla=random.choice(place)
  print("\n",f"{sub} {obj} {pla}")
  user_input1=input("do you like to print more line like this? if yes please type yes otherwise type no ").strip().lower()
  if user_input1== "no":
    break
  elif user_input1== "yes":
    try:
      User_input2=int(input("how many lines do you want to print="))
      for i in range(User_input2):
        sub=random.choice(subject)
        obj=random.choice(Object)
        pla=random.choice(place)
        print("\n",f"{sub} {obj} {pla}")
    except ValueError:
      print("invalid input")
```
